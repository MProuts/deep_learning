```python
# Book-specific setup
# ===================
! [ -e /content ] && pip install -Uqq fastbook
import fastbook
fastbook.setup_book()
from fastbook import *
from fastai.vision.widgets import *
# Drive-specific setup
# ====================
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once per notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```python
# Download a file based on its file ID.
# A file ID looks like: laggVyWshwcyP6kEI-y_W3P8D26sz
file_id = '1_DL5yE2OPtWQhWKtzklrSpcsNlt7QVeo'
import_file = drive.CreateFile({'id': file_id})
import_file.GetContentFile('import.pkl')
```

```python
learn_inf = load_learner('import.pkl')
```

```python
# create buttons
btn_upload = widgets.FileUpload()
btn_run = widgets.Button(description='Classify')
# create output
out_pl = widgets.Output()
# create label
lbl_pred = widgets.Label()
# create click handler
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
VBox([widgets.Label('Select your dog!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
